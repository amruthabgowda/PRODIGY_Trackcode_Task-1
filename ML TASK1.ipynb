```python
# Importing necessary libraries
import numpy as np
import pandas as pd

# Generating a larger dataset with 100 rows
np.random.seed(42)  # For reproducibility

# Random data generation
square_footage = np.random.randint(800, 5000, 100)  # Random square footage between 800 and 5000
bedrooms = np.random.randint(1, 6, 100)  # Random number of bedrooms between 1 and 5
bathrooms = np.random.randint(1, 4, 100)  # Random number of bathrooms between 1 and 3
price = square_footage * 150 + bedrooms * 30000 + bathrooms * 50000 + np.random.randint(-50000, 50000, 100)

# Create a DataFrame
large_df = pd.DataFrame({
    'SquareFootage': square_footage,
    'Bedrooms': bedrooms,
    'Bathrooms': bathrooms,
    'Price': price
})

# Display the first 20 rows of the dataset
print(large_df.head(20))

# Optional: Display the full dataset in tabular format
large_df
```
